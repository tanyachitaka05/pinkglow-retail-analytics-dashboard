```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
```

```python
# Load dataset
# Make sure your CSV file is in the same folder
# Replace filename if needed

df = pd.read_csv("cleaned_sales_data.csv")

# Convert date column
# This helps create monthly trends

df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
# Revenue by category
fig_category = px.bar(
    df.groupby('Product Category')['Total Amount'].sum().reset_index(),
    x='Product Category',
    y='Total Amount',
    color='Product Category',
    title=' Revenue by Product Category'
)


# Gender distribution
fig_gender = px.pie(
    df,
    names='Gender',
    title=' Customer Gender Distribution'
)

# Monthly sales trend
monthly_sales = df.groupby('Month')['Total Amount'].sum().reset_index()

fig_monthly = px.line(
    monthly_sales,
    x='Month',
    y='Total Amount',
    markers=True,
    title=' Monthly Revenue Trend'
)

# Age distribution
fig_age = px.histogram(
    df,
    x='Age',
    nbins=10,
    title=' Customer Age Distribution'
)

# -----------------------------
# Dash App
# -----------------------------

app = Dash(__name__)

app.layout = html.Div(
    style={
        'backgroundColor': '#ffe6f2',
        'padding': '20px',
        'fontFamily': 'Arial'
    },
    children=[

        html.H1(
            ' PinkGlow Retail Analytics Dashboard',
            style={
                'textAlign': 'center',
                'color': '#d63384'
            }
        ),

        html.Div([
            html.Div([
                html.H3('Total Revenue'),
                html.H2(f"R {df['Total Amount'].sum():,.2f}")
            ], style={
                'backgroundColor': 'white',
                'padding': '20px',
                'borderRadius': '15px',
                'width': '30%',
                'display': 'inline-block',
                'margin': '10px',
                'textAlign': 'center'
            }),

            html.Div([
                html.H3('Total Customers'),
                html.H2(df['Customer ID'].nunique())
            ], style={
                'backgroundColor': 'white',
                'padding': '20px',
                'borderRadius': '15px',
                'width': '30%',
                'display': 'inline-block',
                'margin': '10px',
                'textAlign': 'center'
            }),

            html.Div([
                html.H3('Average Purchase'),
                html.H2(f"R {df['Total Amount'].mean():,.2f}")
            ], style={
                'backgroundColor': 'white',
                'padding': '20px',
                'borderRadius': '15px',
                'width': '30%',
                'display': 'inline-block',
                'margin': '10px',
                'textAlign': 'center'
            })
        ]),

        dcc.Graph(figure=fig_category),
        dcc.Graph(figure=fig_gender),
        dcc.Graph(figure=fig_monthly),
        dcc.Graph(figure=fig_age)
    ]
)

# Run app
if __name__ == '__main__':
    app.run(debug=True)

```
